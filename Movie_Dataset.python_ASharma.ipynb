Task 5: In the dataset, there are some movies for which the budget and revenue columns have the value 0, which means 
    unknown values. Remove the rows withvalue 0 from both the budget and revenue columns.

```python
temp_df = movie_df.copy()
temp_df.shape
```

```python
# showing the zero values for budget and revenue
print('Bugdet with 0 value.......................................')
display(movie_df[(movie_df.budget == 0)])
print('Revenue with 0 value...........................................')
display(movie_df[movie_df.revenue == 0])
```

```python
# removing 0's from both the budget and revenue by using ~(tilde)

condition_1 = (movie_df['budget'] == 0)
condition_2 = (movie_df['revenue'] == 0)

combined_condition = condition_1 | condition_2
movie_df = movie_df[~combined_condition] # using ~ tilde func to remove zeros from bugdet and revenue column
movie_df
```

```python
kmeans = KMeans(n_clusters=2)  # choose the number of clusters
movie_df['cluster'] = kmeans.fit_predict(x)
plt.figure(figsize=(10,5))

# Plot data points based on cluster assignments
for cluster_num in movie_df['cluster'].unique():
    cluster_data = movie_df[movie_df['cluster'] == cluster_num]
    plt.scatter(cluster_data['popularity'], cluster_data['runtime'], label=f'Cluster {cluster_num}')

# Add labels and legend
plt.xlabel('Popularity')
plt.ylabel('Runtime')
plt.legend()

plt.show()
```

Task 10: Sort the data in descending order based on revenue and filter the top 500 movies. Find the measures of central tendency for the following columns using the filtered data:

1.Budget
2.Revenue
3.Runtime

Perform outlier analysis for the above three columns using box plots.

Summary Statistics:

For the analysis use of statistics is very helpful in context of Central Tendencies. It can shows us the outliers with calculation of Mean and Median which shows if Mean is higher than Median.
Analysis are made on the filtered data of top 500 movies which shows the outlier of budget , revenue and runtime. It can clearly be seen in the calculation where mean is greater than the median shows outlier. Here I am comparing top_500_movies and the least_500_movies to get the better understanding of budget and revenue

```python
#sorting data in descending order
sorted_movies = movie_df.sort_values(by = 'revenue',ascending = False)
# Filter the top 500 movies.
top_500_movies = sorted_movies[['budget','revenue','runtime']].head(500)
display(top_500_movies)
```

For instance, in a dataset, most the movies might have moderate budget,revenue and runtime , but a few extremely high movies could significantly raise the mean. The mean and median can provide insights into the shape of the data distribution. If the mean is substantially greater than the median, it suggests that there are relatively few very high values in the dataset as seen in the calculation above in both Top and Least movies that shows the difference of Good performing movies and the bad performing movies.

```python
# Calculate the measures of central tendency
budget_mean = top_500_movies['budget'].mean()
budget_median = top_500_movies['budget'].median()
budget_mode = top_500_movies['budget'].mode().values[0]

revenue_mean = top_500_movies['revenue'].mean()
revenue_median = top_500_movies['revenue'].median()
revenue_mode = top_500_movies['revenue'].mode().values[0]

runtime_mean = top_500_movies['runtime'].mean()
runtime_median = top_500_movies['runtime'].median()
runtime_mode = top_500_movies['runtime'].mode().values[0]

# Display the results
print("Measures of Central Tendency for Budget:")
print(f"Mean: {budget_mean}")
print(f"Median: {budget_median}")
print(f"Mode: {budget_mode}")
print(f"..........................................................")

print("\nMeasures of Central Tendency for Revenue:")
print(f"Mean: {revenue_mean}")
print(f"Median: {revenue_median}")
print(f"Mode: {revenue_mode}")
print(f"..........................................................")

print("\nMeasures of Central Tendency for Runtime:")
print(f"Mean: {runtime_mean}")
print(f"Median: {runtime_median}")
print(f"Mode: {runtime_mode}")
```

```python
# Plotting bar graphs for budget, revenue, and runtime
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(5, 10))

# Bar graph for budget
axes[0].bar(top_500_movies.index, top_500_movies['budget'], color='blue', alpha=0.7)
axes[0].set_title('Budget for Top 500 Movies')

# Bar graph for revenue
axes[1].bar(top_500_movies.index, top_500_movies['revenue'], color='green', alpha=0.7)
axes[1].set_title('Revenue for Top 500 Movies')

# Bar graph for runtime
axes[2].bar(top_500_movies.index, top_500_movies['runtime'], color='orange', alpha=0.7)
axes[2].set_title('Runtime for Top 500 Movies')

# Set common labels and layout
for ax in axes:
    ax.set_xlabel('Movie Index')
    ax.set_ylabel('Amount')
    ax.grid(True)

plt.tight_layout()
plt.show()
```

Here in the bar graphs, movie_df sorted out according to top 500 good performing movies on the bases of Budget,Revenue and Runtime. It seems that according to the budget movies are giving revenue and the runtime is also high.
